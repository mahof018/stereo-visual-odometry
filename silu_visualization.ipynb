```python
# Plotten der SiLU gegenüber der ReLU
import math
from matplotlib import pyplot as plt
import numpy as np
def sigmoid(x):  
    return 1 / (1 + np.exp(-x))

def ReLU(x):
    return x * (x > 0)

plt.figure(figsize=(4, 2))

x = np.linspace(-5,5,1000)
beta = 1
swish = x*sigmoid(beta*x)
plt.plot(x,ReLU(x),label="ReLU",color="r")
plt.plot(x,swish,label="SiLU",color="b")
plt.grid()
plt.legend()
plt.xlim(-5,5)
plt.show()
```
